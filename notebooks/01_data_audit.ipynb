```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/Sambhajinagar_Dataset_V3_Final.csv")

print("Dataset loaded successfully")
```

```python
print("Dataset shape:", df.shape)

print("\nColumn names:")
print(df.columns.tolist())

print("\nData types:")
print(df.dtypes)
```

```python
print("Missing values per column:")
print(df.isnull().sum())

print("\nDuplicate rows:")
print(df.duplicated().sum())

print("\nFirst 5 rows:")
display(df.head())
```

```python
landcover_counts = (
    df["LandCover"]
    .value_counts()
    .sort_index()
)

print("LandCover class counts:")
print(landcover_counts)

print("\nNumber of unique LandCover classes:", df["LandCover"].nunique())
```

```python
numeric_summary = df[
    [
        "Elevation",
        "LST",
        "Latitude",
        "Longitude",
        "NDBI",
        "NDVI",
        "Population"
    ]
].describe().T

numeric_summary
```

```python
audit_summary = pd.DataFrame({
    "Metric": [
        "Rows",
        "Columns",
        "Missing Values",
        "Duplicate Rows",
        "Unique LandCover Classes"
    ],
    "Value": [
        df.shape[0],
        df.shape[1],
        int(df.isnull().sum().sum()),
        int(df.duplicated().sum()),
        df["LandCover"].nunique()
    ]
})

audit_summary
```

```python
import os

os.makedirs("../outputs/reports", exist_ok=True)

audit_summary.to_csv(
    "../outputs/reports/data_audit_summary_v3.csv",
    index=False
)

print("Saved: outputs/reports/data_audit_summary_v3.csv")
```

## Dataset V3 Audit Conclusion

Dataset V3 contains 9,893 sampled locations and 10 original columns. No missing values or duplicate rows were found.

The dataset includes seven ESA WorldCover land-cover classes: Tree cover, Shrubland, Grassland, Cropland, Built-up land, Bare/sparse vegetation, and Permanent water bodies.

The `LandCover` field is categorical and will be one-hot encoded during feature engineering. The `system:index` and `.geo` fields will be removed during data cleaning. Latitude and Longitude will be retained only for mapping and spatial block cross-validation, not for model training.
